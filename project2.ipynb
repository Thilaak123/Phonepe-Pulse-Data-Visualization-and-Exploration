**Aggregated** - **Transaction** :

```python
import pandas as pd
import json
import os
# The list of states in india
path="/content/pulse/data/aggregated/transaction/country/india/state/"
agg_state_list1 = os.listdir(path)
agg_state_list1
```

```python
#Extracting the datas to create a dataframe
head ={ 'state':[], 'year':[], 'Quater' :[], 'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in agg_state_list1:
    path_i=path+i+"/"
    agg_year=os.listdir(path_i)
    for j in agg_year:
      path_j=path_i+j+"/"
      agg_year_lst=os.listdir(path_j)
      for k in agg_year_lst:
        path_k=path_j+k
        Data=open(path_k,'r')
        D=json.load(Data)
        for z in D['data']['transactionData']:
          name=z['name']
          count=z['paymentInstruments'][0]['count']
          amount=z['paymentInstruments'][0]['amount']
          head['Transaction_type'].append(name)
          head['Transaction_count'].append(count)
          head['Transaction_amount'].append(amount)
          head['state'].append(i)
          head['year'].append(j)
          head['Quater'].append(int(k.strip('.json')))

agg_trans = pd.DataFrame(head)
agg_trans
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("/content/agg_trans.csv")

fig = px.choropleth(
    df,
    geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
    featureidkey='properties.ST_NM',
    locations='state',
    color='Transaction_count',
    color_continuous_scale='Reds'
)
fig.update_geos(fitbounds="locations", visible=False)
fig.show()
```
